```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from   mpl_toolkits.mplot3d import Axes3D
```

```python
font_size = 14
pd_reader = pd.read_csv("./dat00.csv")
print(pd_reader.head())
time_plt = (pd_reader["timeStamp"]-pd_reader["timeStamp"][0])/1000
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(pd_reader[" xr"], pd_reader[" yr"], pd_reader[" zr"], linestyle='--')
ax.plot3D( pd_reader[" OTx"], pd_reader[" OTy"], pd_reader[" OTz"])
plt.legend(["Ref", "Real"], loc='upper center', ncol= 2, fontsize = font_size)
```

```python

"""
tracking performance
"""
fig = plt.figure()
##
plt.subplot(3,1,1)
plt.plot(time_plt, pd_reader[" xr"], linestyle='--')
plt.plot(time_plt, pd_reader[" OTx"])
plt.ylabel(r'$x~\mathrm{[m]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.grid(color='gray', linestyle=':')
plt.legend(["Ref", "Real"], loc='upper center',bbox_to_anchor=(0.5, 2), ncol= 2, fontsize = font_size)
##
plt.subplot(3,1,2)
plt.plot(time_plt, pd_reader[" yr"], linestyle='--')
plt.plot(time_plt, pd_reader[" OTy"])
plt.ylabel(r'$y~\mathrm{[m]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
# plt.xlim(0, tf+dt)
plt.grid(axis='both',color='gray', linestyle=':')
##
plt.subplot(3,1,3)
plt.plot(time_plt, pd_reader[" zr"], linestyle='--')
plt.plot(time_plt, pd_reader[" OTz"])
plt.xlabel('Time [s]',fontsize = font_size)
plt.ylabel(r'$z~\mathrm{[m]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
# plt.xlim(0, tf+dt)
plt.grid(axis='both',color='gray', linestyle=':')

plt.tight_layout()
fig.align_ylabels()
```

```python
fig = plt.figure()
##
plt.subplot(2,1,1)
plt.plot(time_plt, pd_reader[" roll_ref"], linestyle='--')
plt.plot(time_plt, pd_reader[" roll"])
plt.ylabel(r'$\phi~\mathrm{[deg]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.grid(color='gray', linestyle=':')
plt.legend(["Ref", "Real"], loc='upper center',bbox_to_anchor=(0.5, 2), ncol= 2, fontsize = font_size)
##
plt.subplot(2,1,2)
plt.plot(time_plt, pd_reader[" pitch_ref"], linestyle='--')
plt.plot(time_plt, pd_reader[" pitch"])
plt.ylabel(r'$\theta~\mathrm{[deg]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
# plt.xlim(0, tf+dt)
plt.grid(axis='both',color='gray', linestyle=':')
##


plt.tight_layout()
fig.align_ylabels()
```

```python
"""
tracking performance
"""
fig = plt.figure()
##
plt.subplot(3,1,1)
plt.plot(time_plt, pd_reader[" vex"])
plt.ylabel(r'$v_{ex}~\mathrm{[m/s]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.grid(color='gray', linestyle=':')
##
plt.subplot(3,1,2)
plt.plot(time_plt, pd_reader[" vey"])
plt.ylabel(r'$v_{ey}~\mathrm{[m/s]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.grid(axis='both',color='gray', linestyle=':')
##
plt.subplot(3,1,3)
plt.plot(time_plt, pd_reader[" vez"])
plt.xlabel('Time [s]',fontsize = font_size)
plt.ylabel(r'$v_{ez}~\mathrm{[m/s]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.grid(axis='both',color='gray', linestyle=':')

plt.tight_layout()
fig.align_ylabels()
```
